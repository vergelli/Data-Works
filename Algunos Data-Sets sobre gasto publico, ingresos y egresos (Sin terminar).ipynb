#### Estudio de los valores NaN de las columnas "programa_inversiones_prioritarias" y "resultado_primario_pip".
#### Puede que estos subsidios no hayan sido tenidos en cuenta para los años 2016 y 2017, pues se implementaron los dos subsidios, desde 2018-01-01; Luego,"programa_inversiones_prioritarias" se anula a partir de la fecha 2019-01-01. (Es decir solo se mantuvo un año). Luego, "resultado_primario_pip" sigue vigente hasta la fecha.

#### En consecuencia, vamos a llenar los NaN con ceros.

```python
data_ingresos_gastos_mensual["resultado_primario_pip"]
```

```python
data_ingresos_gastos_mensual["programa_inversiones_prioritarias"] = data_ingresos_gastos_mensual["programa_inversiones_prioritarias"].fillna(0)
data_ingresos_gastos_mensual["resultado_primario_pip"] = data_ingresos_gastos_mensual["resultado_primario_pip"].fillna(0)
```
